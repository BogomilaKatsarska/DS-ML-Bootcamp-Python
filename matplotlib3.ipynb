```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(0, 5, 11)
y = x ** 2

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', linewidth=1, alpha=0.8, linestyle='-', marker='o', markersize=20,
        markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green') #RGB HEX CODE, alpha - controlls the transperancy
ax.set_xlim([0,1])
```
