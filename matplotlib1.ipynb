```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
x = np.linspace(0, 5, 11)
y = x ** 2

#functional method
# '-r' is for colour - as in Matlab
plt.plot(x, y, 'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
# plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, '-r')
plt.subplot(1, 2, 2)
plt.plot(y, x, '-b')
```

```python
# Object Oriented 
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #left, bottom, width and height as params
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.plot(x, y)
axes2.plot(y, x)
axes1.set_title('LARGER PLOT')
axes2.set_title('SMALLER PLOT')
```
